```python
from random import randint
import statistics
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Cчитывание выборки
a = int(input('a = '))
b = int(input('b = '))
f = open('data_Lapshina.txt', 'w')
for i in range (10000):
    f.write(str(randint(a,b)) + ' ')
f.close()
```

```python
# Выборочное среднее

with open('data_Lapshina.txt') as f:
    for line in f:
        my_list = [int(x) for x in line.split()]
# print(my_list)
summ = 0
for el in my_list:
    summ += el 
mean = summ / len(my_list)
print('ВЫБОРОЧНОЕ СРЕДНЕЕ')
print('Результат модуля statistics:',statistics.mean(my_list))
print('Результат функции:', mean)
```

```python
# Мода
freq = {}
for item in my_list:
    try:
        freq[item] = freq[item] + 1
    except Exception:
        freq[item] = 1
pos = max(freq.values())

# Ключ по значению
def get_key(d, value):
    for k, v in d.items():
        if v == value:
            return k
        
moda = get_key(freq,pos)
print('МОДА')
print('Результат модуля statistics:',statistics.mode(my_list))
print('Результат функции:', moda)
```

```python
# Медиана
my_list.sort()
mediana = (my_list[5000] + my_list[5001])/2
print('МЕДИАНА')
print('Результат модуля statistics:',statistics.median_high(my_list))
print('Результат функции:', mediana)
```

```python
# Коэффициент вариации
def variance(my_list):
    mean = statistics.mean(my_list)
    summ = 0
    for i in my_list:
        summ += (i - mean)**2
    disp = summ/len(my_list)
    return disp

koeff_var = (variance(my_list)**(1/2))/mean 
print('КОЭФФИЦИЕНТ ВАРИАЦИИ')
print('Результат встроенных функций:', np.std(my_list)/np.mean(my_list))
print('Результат функции:', koeff_var)
```

```python
# Функция распределения
freq = {}
for item in sorted(my_list):
    try:
        freq[item] = freq[item] + 1
    except Exception:
        freq[item] = 1
x = []
y = []
for c,d in freq.items():
    x.append(c)
    y.append(d)
y = np.array(y)/len(my_list)
plt.figure(figsize=(20,10))
plt.plot(x, y)
plt.show()
```
