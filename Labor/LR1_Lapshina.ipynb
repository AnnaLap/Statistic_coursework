```python
from random import randint
import matplotlib.pyplot as plt

def frequency(a, b, n):
    my_list = []
    for i in range (n):
        my_list.append(randint(a,b))
    k = 0  
    
# 1
    my_dict = {}
    for i in range(a, b+1):
        for item in my_list:
            if i == item:
                k += 1            
        my_dict[i] = k/n
        k = 0
    print(my_dict)
    plt.plot(my_dict.keys(), my_dict.values(), 'bo')
    plt.grid()
    
# 2
    f = {}
    for num in my_list:
        try:
            f[num] += 1
        except Exception:
            f[num] = 1
    list_keys = list(f.keys())
    list_keys.sort()
    for i in list_keys:
        print(i, ':', f[i]/n)
        plt.plot(i, f[i]/n, 'r*')
```

```python
frequency(8, 80, 1000)
```
