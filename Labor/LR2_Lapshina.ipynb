```python
from math import factorial as fc 
n = int(input())
k = int(input())
if n <= k:
    print('Введите другие данные')
else: 
    c = fc(n)/(fc(k)*fc(n - k))
    print(c)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin, cos
def f(x):
    if x <= 0:
        return cos(x)
    else:
        return sin(x) 
x=np.linspace(-10,10,50)
f2 = np.vectorize(f)
y = f2(x)
plt.plot(x, y) 
plt.grid()
```
